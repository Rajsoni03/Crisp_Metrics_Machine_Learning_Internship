#importing Libraries

```python
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Input
from tensorflow.keras.models import Model
import pandas as pd
import numpy as np
```

# Load Data

```python
train_data = pd.read_csv('/content/drive/MyDrive/AI Colab/Internship/Task - 2 (MNIST Dataset)/train.csv')
test_data = pd.read_csv('/content/drive/MyDrive/AI Colab/Internship/Task - 2 (MNIST Dataset)/test.csv')
```

```python
train_data.head()
```

```python
train_img = np.array(train_data.iloc[:,1:]).reshape(-1, 28, 28, 1) / 255
train_labels = np.array(train_data.iloc[:,1])

print(train_img.shape)
print(train_labels.shape)
```

# Create and Train Model

```python
Input_X = Input(shape=(28,28,1))

X = Conv2D(16, (3,3), activation='relu', padding='same')(Input_X)
X = Conv2D(32, (3,3), activation='relu')(X)
X = MaxPool2D((2,2))(X)

X = Conv2D(32, (3,3), activation='relu', padding='same')(X)
X = Conv2D(64, (3,3), activation='relu')(X)
X = MaxPool2D((2,2))(X)

X = Flatten()(X)
X = Dense(64, activation='relu')(X)
X = Dense(10, activation='softmax')(X)

model = Model(inputs=Input_X, outputs=X)
model.summary()
```

```python
model.compile(optimizer='SGD', loss='sparse_categorical_crossentropy', metrics='accuracy')
```

```python
model.fit(train_img,train_labels,
          validation_split=.2,
          batch_size = 64,
          epochs = 2)
```
